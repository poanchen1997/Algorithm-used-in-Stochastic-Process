# Viterbi Algorithm for Hidden Markov model

After calculating the likelihood for the hidden state by forward and backward algorithm, we want to inference the hidden state. This algorithm is used for doing this.

Using the same assumption as previous algorithm !

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data for HMM.csv')
 
V = data['Visible'].values
```

```python
print(V)
```

```python
# Transition Probabilities --> the probability that you will change to next state or stay in the same state
a = np.array(((0.54, 0.46), (0.49, 0.51)))
 
# Emission Probabilities   --> the probability that each hidden state change to each visible state
b = np.array(((0.16, 0.26, 0.58), (0.25, 0.28, 0.47)))
 
# Equal Probabilities for the initial distribution  --> we just assume they are the same for two hidden state
initial_distribution = np.array((0.5, 0.5))
```

## start implement :

```python
def viterbi(V, a, b, initial_distribution):
    ###############################################
    # V = data for visible state                  #
    # a = transition probability                  #
    # b = emission probability                    #
    # initial_distribution = guess of each state  #
    ###############################################
    T = V.shape[0]   # 500
    M = a.shape[0]   # 2
 
    omega = np.zeros((T, M))
    omega[0, :] = np.log(initial_distribution * b[:, V[0]])  # first row   # 1 X 2 * 2 X 1  # take log for easy computing
 
    prev = np.zeros((T - 1, M))                              # use for take the max
 
    for t in range(1, T):
        for j in range(M):   # for two hidden state
            # Same as Forward Probability
            probability = omega[t - 1] + np.log(a[:, j]) + np.log(b[j, V[t]])
 
            # This is our most probable state given previous state at time t (1)
            prev[t - 1, j] = np.argmax(probability)
 
            # This is the probability of the most probable state (2)
            omega[t, j] = np.max(probability)
 
    # Path Array
    S = np.zeros(T)
 
    # Find the most probable last hidden state
    last_state = np.argmax(omega[T - 1, :])
 
    S[0] = last_state
 
    backtrack_index = 1
    for i in range(T - 2, -1, -1):
        S[backtrack_index] = prev[i, int(last_state)]
        last_state = prev[i, int(last_state)]
        backtrack_index += 1
 
    # Flip the path array since we were backtracking
    S = np.flip(S, axis=0)
 
    # Convert numeric values to actual hidden states, we can actually put this part out from the algo !
    result = []
    for s in S:
        if s == 0:
            result.append("A")
        else:
            result.append("B")
 
    return result
```

```python
hidden_inference = viterbi(V, a, b, initial_distribution)
print(hidden_inference)
```

## Let's make a plot and check our answer

```python
c_data = []
for data in hidden_inference:
    if data == "B":
        c_data.append(0.5)
    elif data == "A":
        c_data.append(1.5)
```

```python
plt.figure(figsize=(20,5))
plt.plot(V)
plt.plot(c_data, 'r-')
plt.show()
```

The blue line is the visible state (0, 1, 2), and the red line is the inference of the hidden state (A, B). I set the hidden state to 1.5 and 0.5, so I can put them in the same plot.

We can see that they follow the same pattern !
