# Metropolis Hasting Algorithm for MCMC

Although rejection sampling is very powerful, it has a limitation that we need to know the MLE  $\hat{\theta}_{MLE}$,
which is often a challenge problem when the likelihood function is non-convex. Moreover, in many cases
such as Bayesian analysis, we may not know the exact value of the density function $p(x)$ but only know the
value up to a constant. 

This function is again used to generate sample from our target function $\pi$.

```python
import numpy as np
import scipy.stats as st
import seaborn as sns
import matplotlib.pyplot as plt
```

## Suppose our target function is :

```python
def pi(x):
    return 0.7*st.norm.pdf(x, 0, 1) + 0.3*st.norm.pdf(x, 5, 1)
```

```python
# show how the distribution look like
x = np.linspace(-3, 8, 1000)
fig, ax = plt.subplots(1, 1)
ax.plot(x, pi(x), 'r-', alpha=0.6, label='Mix-Gaussian distribution')
plt.legend()
plt.show()
```

## Start building function :

```python
def Metropolis_Hasting(pi, x0, N):
    ###################################################################
    # pi : This is the target function that we want to generate from  #
    # pf : proposal function, because I use the N(0, 1) for it and    #
    #      won't change, so I just get rid of this argument.          #
    #      If you want to change, make sure you can get "random"      #
    #      sample from this distribution.                             #
    #      (maybe by rejection sampling?)                             #
    # x0 : x0 will be the first sample. most of the time just 0?      #
    # N  : How many samples you want to generate?                     #
    ###################################################################
    x = x0                                # setting start point
    samples = np.zeros(N)                 # create a list to store result
    
    for i in range(N):
        x_star = x + np.random.normal()   # get the next step, if you want to change the propersal function, change here !!
                   # change to other random sample as you want
        U = np.random.uniform(0, 1)
        a = min(pi(x_star)/pi(x), 1)      # calculate the acceptance rate. Because we assume the proposal 
                                          # function is exchangable, q(i|j) = q(j|i), we can remove it.
        
        if U <= a:
            x = x_star                    # if accept, x move to next step, else keep the same position
        samples[i] = x
    
    return samples
```

```python
test = Metropolis_Hasting(pi, 0, 10000)
#print(test)
```

```python
sns.distplot(test)
plt.show()
```

We can see the sample histogram is almost the same with our target function.

Also, we can check the line plot,

```python
plt.plot(test)
plt.show()
```

We can observe that most of the time x stay in 0 and 5. This meet our expection.
