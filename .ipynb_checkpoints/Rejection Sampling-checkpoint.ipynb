# Rejection Sampling:
This function is use to generate random sample from an unknown distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
```

If we want to sample from a distribution, but we don't have any package or function to generate them.

We use "Rejection Sampling" to sample from those distribution.

```python
# imagine this is the function we want to sample from
def f(x):
    return 1.2 - x**4

# for the blue line
xs = np.linspace(-1, 1, 1000)
ys = f(xs)

# plot setting
plt.plot(xs, ys, label="Function used") 
plt.fill_between(xs, ys, 0, alpha=0.2) # fill the region
#plt.xlim(0, 1)                         # limit the x region
plt.ylim(0, 1.25)                      # limit the y region 
plt.xlabel("x")                        # set the name of x-axis
plt.ylabel("y")                        # set the name of y-axis
plt.legend()                           # add the legend
plt.show()
```

Reference: *https://cosmiccoding.com.au/tutorials/rejection_sampling* 

## Start implement:

Use a distribution that we have function to generate the random sample.

```python
def q(x):
    # use standard normal distribution as a handy distribution
    # loc = mean
    # scale = standard deviation
    return st.norm.pdf(x, loc = 0, scale = 1)
```

```python
def rejection_sampling(p, q, M, n):
    #####################################################
    # p is the function that we want to sample from     #
    # q is the function that we know how to sample from #
    # M is the supreme of p(x)/q(x)                     #
    # n is how many samples we want to generate         #
    #####################################################
    
    sample_lst = []       # create a list to store the result
    
    while len(sample_lst) <= n :
        Y = np.random.normal(0, 1)  
        U = np.random.uniform(0, 1)
        
        if U <= (p(Y)/(M*q(Y))):
            sample_lst.append(Y)
            
    return sample_lst
```

```python
x = np.arange(-1.1, 1.1, 0.01)
M = max(f(x)/q(x))
# print(M)
test = rejection_sampling(f, q, M, 100000)
# print(test)
```

```python
# Creating histogram
fig, ax = plt.subplots(figsize =(10, 7))
ax.hist(test)
 
# Show plot
plt.show()
```

```python
# using seaborn to draw the line and histogram
sns.distplot(test)
plt.show()
```

### Conclusion:
we can see that the histogram of the sample we generate is almost the same with the plot of the function "f".

Therefore, we can know our function is success.
