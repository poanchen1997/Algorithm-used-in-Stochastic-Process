# Gibbs sampler :

Gibbs sampling is an alternative approach to sample from a target PDF/PMF based on the idea of MCMC
when the target PDF/PMF is multivariate. The appealing feature of Gibbs sampling is that we only need
to know the conditional PDF/PMF of the target rather than the joint PDF/PMF. But we do need to know
how to sample from the conditional PDF/PMF of each variable given the others.


```python
import numpy as np
import scipy.stats as st
import seaborn as sns
import matplotlib.pyplot as plt
```

Suppose our target distribution is a bivariate normal distribution.

```python
mus = np.array([5, 5])                     # mean
sigmas = np.array([[1, .9], [.9, 1]])      # covariance
```

```python
def p(x, y):
    return st.multivariate_normal.pdf([x, y], mean=mus, cov=sigmas)
```

```python
# show how the distribution look like
from scipy.stats import multivariate_normal

# for line plot
x = np.linspace(0, 10, 50, endpoint=False)
y = multivariate_normal.pdf(x, mean=5, cov=0.9)

fig1 = plt.figure()
ax = fig1.add_subplot(111)
ax.plot(x, y)

# for contour plot
x, y = np.mgrid[1:9:.01, 1:9:.01]
pos = np.dstack((x, y))
rv = multivariate_normal([5, 5], [[1, 0.9], [0.9, 1]])
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.contourf(x, y, rv.pdf(pos))
plt.show()
```

## Start implement : 

```python
# use to sample from the conditional distribution
def conditional_sampler(sampling_index, current_x, mean, cov):
    conditioned_index = 1 - sampling_index 
    # The above line works because we only have 2 variables, u & v
    a = cov[sampling_index, sampling_index]
    b = cov[sampling_index, conditioned_index]
    c = cov[conditioned_index, conditioned_index]
  
    mu = mean[sampling_index] + (b * (current_x[conditioned_index] - mean[conditioned_index]))/c
    sigma = np.sqrt(a-(b**2)/c)
    new_x = np.copy(current_x)
    new_x[sampling_index] = np.random.randn()*sigma + mu
    return new_x
```

The above conditional distribution will change if your target function is different.

For instance, our target function is multivariate normal function, more specifically, it's bivariate normal distribution, and we know that for bivariate normal distribution, it's conditional distribution will also be normal and it's form is fixed. 

However, if your target pdf is not "pretty" enough, finding a way to get samples from conditional distribution will be hard.

```python
def gibbs_sampler(initial_point, num_samples, mean, cov):
    ##################################################################
    # initial_point : the starting point                             #
    # num_samples : we will overall generate num_samples samples.    #
    # mean, cov : use for calculating the conditional distribution.  # 
    ##################################################################
    point = np.array(initial_point)         # starting point: X_0
    samples = np.empty([num_samples, 2])    # sampled points, a num_sample X 2 matrix
    samples[0] = point                      # X_0 is the initial_point

    for i in range(num_samples-1):
        # Sample from p(x_0|x_1)
        point = conditional_sampler(0, point, mean, cov)
        
        # Sample from p(x_1|x_0)
        point = conditional_sampler(1, point, mean, cov)
        samples[i+1] = point

    return samples
```

Let's check our result, 

```python
samples = gibbs_sampler([5.0, 5.0], 1000, mus, sigmas)
```

```python
x = [i[0] for i in samples]
y = [i[1] for i in samples]
plt.scatter(x, y)
plt.show()
```

It's almost the same with our plot of target pdf.
